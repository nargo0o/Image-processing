```
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

1. С помощью библиотеки OpenCV считайте изображение:

```
img = cv2.imread('3.4.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

2. Преобразуйте изображение в палитру RGB и примените гауссово ядро, используя метод cv2.GaussianBlur(), размер ядра (11, 3)

3. Получите изображение с деталями.

```
kernel = np.ones((3, 3), np.float32) / 9
RGB_img_new = cv2.filter2D(RGB_img, -1, kernel)
plt.imshow(RGB_img_new)
```

```
kernel = np.ones((9, 9), np.float32) / 81
RGB_img_new = cv2.filter2D(RGB_img, -1, kernel)
plt.imshow(RGB_img_new)
```

```
kernel = np.ones((3, 3), np.float32) / 16
kernel[0][1] = kernel[1][0] =  kernel[1][2] = kernel[2][1] = 2/16
kernel[1][1] = 4/16
RGB_img_new = cv2.filter2D(RGB_img, -1, kernel)
plt.imshow(RGB_img_new)
```

```
RGB_img_new = cv2.GaussianBlur(RGB_img, (11,3), 3)
plt.imshow(RGB_img_new)
```

```
RGB_img_remaining = RGB_img - RGB_img_new
plt.imshow(RGB_img_remaining)
```
